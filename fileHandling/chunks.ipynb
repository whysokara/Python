```python
big_L = ['hello world\n' for i in range(1000)]

with open('big.txt', 'w') as f:
    f.writelines(big_L)

with open('big.txt', 'r') as f:
    chunk_size = 10
    while True:
        lines = [f.readline() for _ in range(chunk_size)]
        if not lines[0]:
            break
        for line in lines:
            if line:
                print(line, end='')
        print("\n--- End of Chunk ---\n")

```
